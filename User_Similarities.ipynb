# 3. sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of users. pairwise_distances() takes a metric parameter for what distance measure to use. Use cosine similarity for finding similarity among users. Use the following packages.


# 4.from sklearn.metrics import pairwise_distances

# 5.from scipy.spatial.distance import cosine, correlation

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
pivot_table = df_ratings.pivot_table(index='userId', columns='movieId', values='rating', fill_value=0)
user_similarity = 1 - pairwise_distances(pivot_table, metric='cosine')
user_similarity_df = pd.DataFrame(user_similarity, index=pivot_table.index, columns=pivot_table.index)
print(user_similarity_df)
```

# 6. Find the 5 most similar user for user with user Id 10.

```python
user_similarity_10 = user_similarity_df.loc[10]
most_similar_users = user_similarity_10.sort_values(ascending=False)
top_5_similar_users = most_similar_users[1:6]
print("Top 5 most similar users for user 10:")
print(top_5_similar_users)
```

# 7. Use the “movies” dataset to find out the names of movies, user 2 and user 338 have watched in common and how they have rated each one of them.

```python
moviesub=set(df_ratings.loc[df_ratings.userId==1, 'movieId']).intersection(set(df_ratings.loc[df_ratings.userId==338, 'movieId']))
for i in moviesub:
    print(df_movies.loc[df_movies.movieId==i, 'title'].values[0])
```

```python
for i in moviesub:
    print('Movie ID', i, ": ", df_movies.loc[df_movies.movieId == i, 'title'].values[0])
    print('Rating by User 1:', df_ratings.loc[((df_ratings.userId == 1) & (df_ratings.movieId == i)), 'rating'].values[0])
    print('Rating by User 338:', df_ratings.loc[((df_ratings.userId == 338) & (df_ratings.movieId == i)), 'rating'].values[0])
```

# 10. Find the top 5 movies which are similar to the movie “Godfather”.

```python
godfather_movie_id_array = df_movies.loc[df_movies['title'] == 'Godfather', 'movieId'].values

# Check for any matches
if len(godfather_movie_id_array) > 0:
    godfather_movie_id = int(godfather_movie_id_array[0])
    godfather_similarity = movie_similarity_df[godfather_movie_id].sort_values(ascending=False)
    top_similar_movies = godfather_similarity.drop(godfather_movie_id).head(5)
    top_similar_movies_df = pd.merge(
        pd.DataFrame(top_similar_movies, columns=['Similarity']),
        df_movies[['movieId', 'title', 'genres']],
        left_index=True,
        right_index=True
    )
    print("Top 5 Movies Similar to Godfather:")
    print(top_similar_movies_df)
else:
    print("No movie with the title 'Godfather' found.")
```
